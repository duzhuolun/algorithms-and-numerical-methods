```
import numpy as np

def lab07_rsa_duzhuolu(p,q,m):
  N = p*q
  theta = (p-1)*(q-1)

  e = 0
  d = 0

  for i in range(2,theta):
    if np.gcd(i,theta) == 1:
      e = i
      break

  for i in range(e*theta+1):
    if (i*e)%theta == 1:
      d = i
      break

  m_num = np.array([ord(c) for c in m])

  c = np.zeros(len(m_num),dtype=int)
  m_d_num = np.zeros(len(m_num),dtype=int)

  for i, n in enumerate(m_num):
    c[i] = pow(int(n),e,N)
    m_d_num[i] = pow(int(c[i]),d,N)

  m_d = ''.join(chr(i) for i in m_d_num)
    
  return e, d, c, m_d
```
