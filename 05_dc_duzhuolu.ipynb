```
import numpy as np
import matplotlib.pyplot as plt
import time
```

```
def lab05_dc_duzhuolu(array):
  l = len(array)
  if l > 5:
    idx = int(np.floor(l/2))
    if(array[idx]>array[idx-1]):
      return lab05_dc_duzhuolu(array[idx:])
    else:
      return lab05_dc_duzhuolu(array[:idx])
  else:
    return array.max()
```

```
def PlotTimes_lab05_duzhuolu(n):
  size_arr = np.array(range(5,n+1))
  time_arr = np.zeros(n-4)

  for i in range(5,n+1):
    m = np.random.randint(1,i)
    f = np.sort(np.random.rand(m))
    s = np.sort(np.random.rand(i-m))[::-1]
    array = np.concatenate((f, s))
    start = time.clock()
    lab05_dc_duzhuolu(array)
    t = time.clock() - start    
    time_arr[i-5] = t 

  plt.scatter(size_arr,time_arr)
  plt.xlabel('size')
  plt.ylabel('time')
  plt.show()
```
