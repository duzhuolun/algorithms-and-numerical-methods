```
import numpy as np
import matplotlib.pyplot as plt
import time
```

```
def lab04_sorting_duzhuolu(a):
  t = time.clock()
  n = len(a)
  for i in range(0, n):
    changed = False
    for j in range(0,n-i-1):
      if a[j] > a[j+1]:
        b = a[j]
        a[j] = a[j+1]
        a[j+1] = b
        changed = True
    if not changed:
      break
  return n, a, (time.clock() - t)
```

```
def lab04_PlotTimes_duzhuolu(n):
  mean_times, max_times = np.zeros(n-1), np.zeros(n-1)
  for i in range(n-1):
    times_n = np.zeros(100)
    for j in range(100):
      unsort = np.random.rand(i+2)
      n, a, t = lab04_sorting_duzhuolu(unsort)
      times_n[j] = t
    mean_times[i] = np.mean(times_n)
    max_times[i] = np.max(times_n)
  p1 = plt.scatter(range(2, n+1),mean_times)
  p2 = plt.scatter(range(2, n+1),max_times)
  plt.legend((p1,p2), ('mean time', 'max time'))
```
