```
import numpy as np
import time
```

```
def lab06_hash_duzhuolu(n,K,check):
  np.random.seed(2)
  startTime = time.perf_counter()

  B = []
  H = []
  a = pow(2,n-np.arange(1,n+1))
  
  for k in range(K):
    x_hat = np.random.randint(2, size = n)

    if check == 1:
      h = sum(a*x_hat)
      if Check(h,H) == False:
        H.append(h)
        B.append(x_hat)
    
    if check == 2:
      if Check2(x_hat,B) == False:
        B.append(x_hat)

  return len(B), time.perf_counter() - startTime
```

```
def Check(h,H):
  if (np.array(H) == h).any():
    return True
  else:
    return False
```

```
def Check2(x,B):
  for b in B:
    if (b==x).all():
      return True
  return False
```
