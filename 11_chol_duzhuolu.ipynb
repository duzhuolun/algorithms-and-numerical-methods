```python
import numpy as np
import time

def cholesky(A):
  n = len(A)
  L = np.zeros(A.shape)

  for i in range(n):
    for k in range(i+1):
      sum_t = sum(L[i][j] * L[k][j] for j in range(k))
      if (i == k):
          L[i][k] = np.sqrt(A[i][i] - sum_t)
      else:
          L[i][k] = (1.0 / L[k][k] * (A[i][k] - sum_t))
  return L

def lab11_chol_duzhuolu(arrayA):
  
  start = time.process_time()
  myL = cholesky(arrayA)
  time1 = time.process_time() - start

  start = time.process_time()
  pyL = np.linalg.cholesky(arrayA)
  time2 = time.process_time() - start

  return myL, pyL, time1, time2
```
