```
import time

def lab02_recursion_duzhuolu(n):
  t = time.clock()
  f1 = fib1(n)
  t1 = time.clock() - t
  t = time.clock();
  f2 = fib2(n)
  t2 = time.clock() - t
  return f1,f2,t1,t2

def fib1(n):
  if n < 0:
    return 'error'
  elif n < 2: 
    return n
  else:
    return fib1(n-1) + fib1(n-2)

def fib2(n):
  if n < 0:
    return 'error'
  elif n < 2:
    return n
  else:
    a = [0] * (n+1)
    a[1] = 1
    for i in range(2,n+1):
      a[i] = a[i-1] + a[i-2]
    return a[n]
```

```
import matplotlib.pyplot as plt

def lab02_PlotTimes_duzhuolu(n):
  T1 = []
  T2 = []
  for i in range(1,n+1):
    f1,f2,t1,t2 = lab02_recursion_duzhuolu(i)
    T1.append(t1)
    T2.append(t2)
  plt.scatter(T1, T2)
```
